```python
import geopandas as gpd
import requests as req
```

```python
layer = 'https://oaiee.scj.gov.co/agc/rest/services/Tematicos_NR/EquipamientoPMSDSCJ/MapServer/25'
fields_route = f'{layer}/query?where=1=1&outFields=*&f=geojson'
metadata_route = f'{layer}?f=pjson'
```

```python
metadata = req.get(metadata_route).json()
```

```python
EPSG = metadata.get('sourceSpatialReference').get('latestWkid')
```

```python
CP = gpd.read_file(fields_route)
CP = CP.to_crs(epsg = EPSG)
#CP
```

```python
CP.plot(color='none', edgecolor='black')
```
